```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)


tienda1.head(20)
#tienda.info()
#tienda.describe()
```

```python
print(f"Numero total de precios:", len(tienda1))
print(tienda1['Costo de envío'].head())
print("\n")
print(tienda2['Producto'].head())
print("\n")
print(tienda3['Producto'].head())
print("\n")
print(tienda4['Producto'].head())
```

```python
# @title Analisis de facturacion

tiendas = [tienda1, tienda2, tienda3, tienda4] # creo una lista donde contiene a las columnas de cada tienda
nombres = ["tienda1", "tienda2", "tienda3", "tienda4"]

for i in range(len(tiendas)):
  total = 0
  for precio in tiendas[i]["Precio"]:
      total += precio
  total = total / 100
  print(f"El ingreso total de {nombres[i]} es: ${total:,.2f}")
```

```python
# @title (Grafico) Analisis de facturacion
# Grafico(Barras)

tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres = ["tienda1", "tienda2", "tienda3", "tienda4"]

# Calcular ingresos totales y agregarlos a una lista vacia
totales = []
for tienda in tiendas:
  total = tienda["Precio"].sum() / 100
  totales.append(total)

# Crear figura con 3 subgráficos (1 fila, 3 columnas)
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle(f"Análisis de facturacion", fontsize=16)

# === Barras ===

axs[0].bar(nombres, totales, color = "Skyblue") #(eje X, eje Y, color)
axs[0].set_title("Barras") # titulo del grafico
axs[0].set_xlabel("Tienditas") # etiqueta eje X
axs[0].set_ylabel("Ingreso Total de las tienditas") # etiqueta eje Y
#axs[0].tick_params(axis='x', rotacion=45) #(no se porque da error con esta linea)

# === Pastel ===
axs[1].pie(totales,
        labels=nombres,
        autopct="%1.1f%%",
        startangle=90,
        colors=plt.cm.Pastel1.colors
        ) # rotacion inicial para que empiece desde arriba (startangle), autopct muestra el porcentaje con un decimal
axs[1].set_title("Pastel")
axs[1].axis("equal") # hace que el pastel sea un circulo perfecto

# === Lineas ===
axs[2].plot(totales, nombres, marker = 'o', linestyle='-', color = "Darkgreen")
axs[2].set_title("Lineas")
axs[2].set_xlabel("Ingresos totales por tiendita")
axs[2].set_ylabel("Tienditas")
axs[2].tick_params(axis='x', rotation=45)
axs[2].grid(True)

# === Ajustar Espaciado ===
plt.tight_layout(rect=[0, 0.03, 1, 0.90])
plt.show()
```

```python
 # @title Ventas por categoria

tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres = ["tienda 1", "tienda 2", "tienda 3", "tienda 4"]

for i, tienda in enumerate(tiendas):
  mas_vendido = tienda["Categoría del Producto"].value_counts().head(5) #el metodo value counts cuenta las veces que aparece un producto en una tienda

  print(f"\nLos productos mas vendidos de {nombres[i]}")
  for producto, cantidad in mas_vendido.items():
      print(f"{producto}: {cantidad} ventas")
```

```python
# @title Graficos ventas por categoria

tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres = ["tienda 1", "tienda 2", "tienda 3", "tienda 4"]

for i, tienda in enumerate(tiendas):
    mas_vendido = tienda["Categoría del Producto"].value_counts().head(5)

    productos = mas_vendido.index
    cantidades = mas_vendido.values

# Crear figura con 3 subgráficos (1 fila, 3 columnas)
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle(f"Análisis de productos más vendidos - {nombres[i]}", fontsize=16)

# === Gráfico de Barras ===
axs[0].bar(productos, cantidades, color='skyblue')
axs[0].set_title("Barras")
axs[0].set_xlabel("Producto")
axs[0].set_ylabel("Cantidad")
axs[0].tick_params(axis='x', rotation=45)

# === Gráfico de Pastel ===
axs[1].pie(cantidades, labels=productos, autopct="%1.1f%%", startangle=90)
axs[1].set_title("Pastel")
axs[1].axis("equal")  # Para que el pastel sea un círculo

# === Gráfico de Línea ===
axs[2].plot(productos, cantidades, marker='o', linestyle='-', color='green')
axs[2].set_title("Línea")
axs[2].set_xlabel("Producto")
axs[2].set_ylabel("Cantidad")
axs[2].tick_params(axis='x', rotation=45)
axs[2].grid(True)

# Ajustar espaciado
plt.tight_layout(rect=[0, 0.03, 1, 0.90])
plt.show()
```

```python
# @title Productos mas y menos vendidos


for i, tienda in enumerate(tiendas):
    conteo_ventas = tienda["Categoría del Producto"].value_counts()
    mas = conteo_ventas.nlargest(5)
    #mas_vendido = mas.head(5)
    menos_vendido = conteo_ventas.nsmallest(5)

    for producto, cantidad in mas_vendido.items():
        print(f'\nLos productos mas vendidos de {nombres[i]}:\n{producto} : {cantidad}')

    for producto, cantidad in menos_vendido.items():
        print(f'\nLos productos menos vendidos de {nombres[i]}:\n{producto} : {cantidad}')



```

```python
# @title (Graficos) Ventas productos mas o menos vendidos

for tienda in tiendas:
  conteo_ventas = tienda["Categoría del Producto"].value_counts()
  mas_vendido = conteo_ventas.nlargest(5)
  menos_vendido = conteo_ventas.nsmallest(5)


fig, axs = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle(f"Análisis de ventas para {nombres[i]}", fontsize=16)

# === Gráfico de Barras ===
axs[0].bar(mas_vendido.index, mas_vendido.values, color='skyblue')
axs[0].set_title("Más Vendidos - Barra")
axs[0].set_xlabel("Producto")
axs[0].set_ylabel("Cantidad")
axs[0].tick_params(axis='x', rotation=45)

# === Gráfico de Pastel ===
axs[1].pie(mas_vendido.values, labels=mas_vendido.index, autopct="%1.1f%%", startangle=90)
axs[1].set_title("Más Vendidos - Pastel")
axs[1].axis("equal")

# === Grafico de Línea ===
axs[2].plot(menos_vendido.index, menos_vendido.values, marker='o', linestyle='-', color='green')
axs[2].set_title("Menos Vendidos - Línea")
axs[2].set_xlabel("Producto")
axs[2].set_ylabel("Cantidad")
axs[2].tick_params(axis='x', rotation=45)
axs[2].grid(True)

# Ajustar espaciado
plt.tight_layout(rect=[0, 0.03, 1, 0.90])
plt.show()








```

```python
#@title Valoracion media por tienda

for i, tienda in enumerate(tiendas):
    conteo_calificacion = tienda["Calificación"].mean()
    print(f"El promedio de la {nombres[i]} es de: {conteo_calificacion:,.2f}")

    if conteo_calificacion >= 4.0:
      print("Los clientes estan satisfechos con la mayoria de los productos vendidos\n")

    elif conteo_calificacion < 4.0 or conteo_calificacion >= 3.0:
      print("Los clientes estan conformes con parte de los productos vendidos\n")

    elif conteo_calificacion < 3.0:
      print("Los clientes no estan conformes con los productos vendidos\n")
```

```python
#@title (Graficos) Valoracion media por tienda

tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]


conteo = []
niveles_satisfaccion = []

for tienda in tiendas:
  conteo_calificacion = tienda['Calificación'].mean()
  conteo.append(conteo_calificacion)

if conteo_calificacion >= 4.0:
  niveles_satisfaccion.append("Satisfechos")

elif conteo_calificacion >= 3.0:
  niveles_satisfaccion.append("Conformes")

else:
  niveles_satisfaccion.append("inconformes")


# === Graficos ===

fig, axs = plt.subplots(1, 3, figsize  = (18, 5))
fig.suptitle(f"Analisis de valoracion media por tienda", fontsize=16)

# === Grafico de Barras ===
axs[0].bar(nombres, conteo, color='Skyblue')
axs[0].set_title("Barras")
axs[0].set_xlabel("Tiendas")
axs[0].set_ylabel("Promedios de Calififacion")
axs[0].tick_params(axis='x', rotation=45)

# === Grafico de Pastel ===
axs[1].pie(conteo, labels=nombres, autopct="%1.1f%%", startangle=90)
axs[1].set_title("Pastel")
axs[1].axis("equal")  # Para que el pastel sea un círculo

# === Grafico de Linea ===
axs[2].plot(nombres, conteo, marker='o', linestyle='-', color='green')
axs[2].set_title("Linea")
axs[2].set_xlabel("Satisfaccion")
axs[2].set_ylabel("Promedios de Calificacion")
axs[2].tick_params(axis='x', rotation=45)
axs[2].grid(True)

# Ajustar espaciado
plt.tight_layout(rect=[0, 0.03, 1, 0.90])
plt.show()

```

```python
# @title Valoracion envio promedio de la tienda

for tienda in tiendas:
  conteo_envios = tienda["Costo de envío"].mean()
  print(f"El gasto en promedio de envios de la {nombres[i]} es de: ${conteo_envios:,.0f}\n")
```

```python
# @title (Graficos) Envio Promedio de la tienda


costos_promedio = []

for i, tienda in enumerate(tiendas):
  promedio = tienda["Costo de envío"].mean()
  costos_promedio.append(promedio)

# Crear gráficos
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Análisis del Costo Promedio de Envío por Tienda", fontsize=16)

# === Gráfico de Barras ===
axs[0].bar(nombres, costos_promedio, color='skyblue')
axs[0].set_title("Barras")
axs[0].set_xlabel("Tiendas")
axs[0].set_ylabel("Costo Promedio (COP)")
axs[0].tick_params(axis='x', rotation=45)

for i, valor in enumerate(costos_promedio):
    axs[0].text(i, valor + 500, f"${valor:,.0f}", ha='center', va='bottom', fontsize=10)

# === Gráfico de Pastel ===
axs[1].pie(costos_promedio, labels=nombres, autopct="%1.1f%%", startangle=90)
axs[1].set_title("Pastel")
axs[1].axis("equal")

# === Gráfico de Línea ===
axs[2].plot(nombres, costos_promedio, marker='o', linestyle='-', color='green')
axs[2].set_title("Línea")
axs[2].set_xlabel("Tiendas")
axs[2].set_ylabel("Costo Promedio")
axs[2].tick_params(axis='x', rotation=45)
axs[2].grid(True)

# Ajustar diseño
plt.tight_layout(rect=[0, 0.03, 1, 0.90])
plt.show()





```
